```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import iqr
```

```python
df = pd.read_csv('avocado.csv')
df
```

```python
# Get average daily price of avocados for entire dataset
avg_price = df['AveragePrice'].mean()

# Get median of average daily price
med_price = df['AveragePrice'].median()

'''
`avg_price` is the average of the average daily price across the dataset  
`med_price` is the median of the average daily price across the dataset
'''

# Get a list of column names
my_columns = df.columns.tolist()

# Get min and max values from 'Total Volume'
min_vol = df['Total Volume'].min()
max_vol = df['Total Volume'].max()


# Calculate volume range
range_vol = max_vol-min_vol

print(min_vol, max_vol)
print(range_vol)

'''
The max/min_vol values represent the highest and lowest values in the 'Total Volume' of avocados sold.
The range_vol represents the total spread of the 'Total Volume'
'''
```

```python
# Calculate min, max, and range for average daily prices
min_price = df['AveragePrice'].min()
max_price = df['AveragePrice'].max()
range_price = max_price-min_price

print(min_price, max_price)
print(range_price)

'''
We've determined the values of the min and max of the range 
as well as the spread for the daily average price of avocados
'''
```

```python
plt.hist(df['AveragePrice'], bins=14, range=(0,3.5))
plt.title('AveragePrice')
plt.show()

plt.hist(df['Total Volume'], bins=10, range=(0,400000))
plt.title('Total Volume')
plt.show()

'''
The AveragePrice is following a normal distribution. 
The Total Volume is right skewed
'''
```

```python
# Calculate the interquartile range of the price
iqr_price = iqr(df['AveragePrice'], rng=(25,75))
iqr_vol = iqr(df['Total Volume'])

print(iqr_price, iqr_vol)
'''
The interquartile range is smaller because the data is right skewed. 
The interquartile range show the range between the 25th and 75th percentile. 
'''
```

```python
# Plot AveragePrice and 'Total Volume'
plt.scatter(df['AveragePrice'], df['Total Volume'])
plt.title('AveragePrice vs Total Volume')
plt.xlabel('Daily Average Price')
plt.ylabel('Total Volume')
plt.show()

# Calculate Pearson correlation between AveragePrice and 'Total Volume'
corr_price_vol = stats.pearsonr(df['AveragePrice'], df['Total Volume'])
print(corr_price_vol)

'''
The plot and pearson value show that the features are not linearly correlated.
There does seem to be some clustering which could afford further analysis. 
'''
```
